```python
import numpy as np
import pandas as pd
import matplotlib . pyplot as plt
from sklearn . tree import DecisionTreeRegressor
from sklearn . ensemble import RandomForestRegressor
from sklearn . preprocessing import OrdinalEncoder , StandardScaler
from sklearn . model_selection import train_test_split
from sklearn . metrics import mean_absolute_error , mean_squared_error
```

```python
dataset_path = '/content/Housing.csv'
df = pd. read_csv ( dataset_path )
df
```

```python
# Lấy danh sách các cột có kiểu dữ liệu là Object (chuỗi ký tự)
categorical_cols = df.select_dtypes(include=['object']).columns.to_list()

# In ra danh sách các cột dạng categorical
print(categorical_cols)
```

```python
# Giả sử df là DataFrame ban đầu của bạn
# Khởi tạo OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

# Mã hóa các cột categorical đã xác định
encoded_categorical_cols = ordinal_encoder.fit_transform(df[categorical_cols])

# Chuyển đổi các cột đã mã hóa thành DataFrame mới
encoded_categorical_df = pd.DataFrame(
    encoded_categorical_cols,
    columns=categorical_cols
)

# Loại bỏ các cột categorical ban đầu khỏi DataFrame chính
numerical_df = df.drop(categorical_cols, axis=1)

# Kết hợp lại DataFrame với các cột đã được mã hóa
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis=1)

# Hiển thị DataFrame mới đã được mã hóa
print(encoded_df.head())
```

```python
# Chuẩn hóa dữ liệu
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)

# Hiển thị dữ liệu đã chuẩn hóa
print(dataset_arr)

# Tách dữ liệu X và y
# Giả sử cột đầu tiên (price) là nhãn (y) và các cột còn lại là đặc trưng (X)
X = dataset_arr[:, 1:]  # Tất cả các cột từ thứ 2 trở đi
y = dataset_arr[:, 0]   # Cột đầu tiên (price)
```

```python
test_size = 0.3
random_state = 1
is_shuffle = True
X_train , X_val , y_train , y_val = train_test_split (
X, y,
test_size = test_size ,
random_state = random_state ,
shuffle = is_shuffle
)
```

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Huấn luyện mô hình Random Forest
regressor = RandomForestRegressor(random_state=42)
regressor.fit(X_train, y_train)

# Huấn luyện mô hình AdaBoost
regressor = AdaBoostRegressor(random_state=42)
regressor.fit(X_train, y_train)

# Huấn luyện mô hình Gradient Boosting
regressor = GradientBoostingRegressor(random_state=42)
regressor.fit(X_train, y_train)

# Đánh giá mô hình
y_pred = regressor.predict(X_val)

# Tính toán MAE và MSE
mae = mean_absolute_error(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

# In kết quả đánh giá
print('Evaluation results on validation set:')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
```
